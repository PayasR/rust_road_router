```python
import numpy as np
import matplotlib as mpl
import seaborn as sns
%matplotlib inline

import pandas as pd

import json
```

# Basic Precustomization

```python
raw = json.load(open("/amd.home/home/i11/zeitz/experiments/tdcch_preliminary/2019-01-23T09:44:44+01:00.json"))
tdcch_queries = pd.DataFrame.from_records([algo for algo in raw['algo_runs'] if algo['algo'] == "Floating TDCCH Query"])
rank_queries = tdcch_queries.loc[lambda q: q['rank'].notna()]
uniform_queries = tdcch_queries.loc[lambda q: q['rank'].isna()]
```

```python
rank_queries.groupby('rank')['num_meeting_nodes', 'num_nodes_in_elimination_tree_search_space',
       'num_relaxed_elimination_tree_arcs', 'num_relaxed_shortcut_arcs', 'running_time_ms', 'unpacking_running_time_ms'].mean()
```

```python
uniform_queries[['num_meeting_nodes', 'num_nodes_in_elimination_tree_search_space',
       'num_relaxed_elimination_tree_arcs', 'num_relaxed_shortcut_arcs', 'running_time_ms']].mean()
```
